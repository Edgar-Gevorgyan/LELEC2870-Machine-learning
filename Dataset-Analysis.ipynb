```python
words_number = ['n_tokens_title', 'n_tokens_content', 'average_token_length'] # NUMBER
words_ratio = ['n_unique_tokens', 'n_non_stop_words', 'n_non_stop_unique_tokens'] # RATIO 
words = words_number + words_ratio

links = ['num_hrefs', 'num_self_hrefs', 'self_reference_min_shares', 'self_reference_max_shares',
         'self_reference_avg_sharess']  # NUMBER


digital_media = ['num_imgs', 'num_videos']  # NUMBER

time_day = ['weekday_is_monday', 'weekday_is_tuesday', 'weekday_is_wednesday', 'weekday_is_thursday',
            'weekday_is_friday', 'weekday_is_saturday', 'weekday_is_sunday',] # BOOL
time_bool = ['is_weekend']  # BOOL
time = time_day + time_bool

keywords_number = ['num_keywords', 'kw_min_min', 'kw_max_min', 'kw_avg_min', 'kw_min_max',
                   'kw_max_max', 'kw_avg_max', 'kw_min_avg', 'kw_max_avg', 'kw_avg_avg',]  # NUMBER
keywords_bool = ['data_channel_is_lifestyle', 'data_channel_is_entertainment', 'data_channel_is_bus',
                 'data_channel_is_socmed', 'data_channel_is_tech', 'data_channel_is_world', ] # BOOL
keywords = keywords_number + keywords_bool

npl_data = ['LDA_00', 'LDA_01', 'LDA_02', 'LDA_03', 'LDA_04', 
            'title_subjectivity',  'abs_title_subjectivity', 'global_subjectivity',
            'global_sentiment_polarity', 'global_rate_positive_words', 'global_rate_negative_words', 
            'rate_positive_words', 'rate_negative_words',
            'avg_positive_polarity', 'min_positive_polarity', 'max_positive_polarity',
            'avg_negative_polarity', 'min_negative_polarity', 'max_negative_polarity', 
            'title_sentiment_polarity', 'abs_title_sentiment_polarity'] # RATIO 
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

preprocessor = ColumnTransformer(
    [
        ('scaler', StandardScaler(), words_number + links + digital_media + keywords_number)
    ],
    remainder='passthrough'
)

pipeline = Pipeline(
    [
        ('preprocessing', preprocessor),
        ('estimator', Lasso(alpha=0.1))
    ]
)
print(pipeline)
print(Lasso())

model = pipeline
model.fit(X1,Y1)
model.score(X1,Y1)
# 0.022887639608145882
# 0.022286923809750836
```

```python
from sklearn.model_selection import learning_curve

N, train_score, val_score = learning_curve(model, X1, Y1, train_sizes=np.linspace(0.2, 1, 10), cv=10)

print(N)
plt.plot(N, train_score.mean(axis=1), label='train')
plt.plot(N, val_score.mean(axis=1), label='validation')
plt.xlabel('train_sizes')
plt.legend()
```
